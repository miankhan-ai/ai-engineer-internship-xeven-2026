# References Section
## Python Memory Management and Data Types

## 1. Overview
Python manages data using a unified object model:
- Everything is an object
- All objects are stored in heap memory
- Variables are references, not containers

---

## 2. Python Data Types

### Core Types

```python
x = 10        # int
pi = 3.14     # float
flag = True   # bool
name = "AI"   # str
### Collection Types

```python
nums = [1, 2, 3]          # list (mutable)
point = (10, 20)          # tuple (immutable)
user = {"name": "Ali"}    # dict (mutable)
unique = {1, 2, 3}        # set (mutable)
```

## 3. Mutable vs Immutable

### Immutable Types

* int, float, bool, str, tuple
* Cannot be changed after creation

```python
a = "hello"
a = a + " world"  # new object created
```

---

### Mutable Types

* list, dict, set
* Modified in place

```python
lst = [1, 2, 3]
lst.append(4)
```

## 4. Memory Model

### Key Idea

* Variables store references
* Objects live in heap memory

```python
x = 10
y = x
```

## 5. Reference Counting

Python tracks how many references point to an object.

```python
import sys

a = [1, 2, 3]
b = a

sys.getrefcount(a)
```

When reference count becomes 0:

* Object is deleted

---

## 6. Garbage Collection

Handles circular references.

```python
a = []
a.append(a)
```

* Reference counting fails here
* GC removes unreachable cycles

## 7. Stack vs Heap (Conceptual)

## 8. Interning Optimization

Python reuses small immutable objects.

```python
a = 100
b = 100
print(a is b)
```

## 9. Identity vs Equality

```python
a = [1, 2]
b = a

print(a == b)  # True (same value)
print(a is b)  # True (same object)
```

---

## 10. Copying Objects

### Assignment (no copy)

```python
a = [1, 2]
b = a
```

### Shallow copy

```python
b = a.copy()
```

### Deep copy

```python
import copy
b = copy.deepcopy(a)
```

---

## 11. Common Pitfall

### Mutable default argument

```python
def add_item(x, lst=[]):
    lst.append(x)
    return lst

add_item(1)
add_item(2)  # unexpected behavior
```

### Correct version

```python
def add_item(x, lst=None):
    if lst is None:
        lst = []
    lst.append(x)
    return lst
```
